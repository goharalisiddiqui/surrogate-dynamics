```python
import os
import numpy as np
import matplotlib.pyplot as plt 
import MDAnalysis as mda
from MDAnalysis.analysis import align
# import nglview as nv
```

```python
run_folder = '/mnt/projects/sne/Gohar/12.PROJECT_Surrogate_Dynamics/surrogate-dynamics/predict_runs/sd_prediction_edvae_4'

predicted_traj_file = os.path.join(run_folder, 'TFT_predicted.pdb')
ref_structure_file = '/mnt/projects/sne/Gohar/12.PROJECT_Surrogate_Dynamics/surrogate-dynamics/inputs/ala2/ref.pdb'
```

```python
predicted_traj = mda.Universe(predicted_traj_file)
predicted_traj.transfer_to_memory()

ref_frame = mda.Universe(ref_structure_file)
```

```python
align.AlignTraj(predicted_traj,  # trajectory to align
                ref_frame,  # reference
                filename='/mnt/projects/sne/Gohar/12.PROJECT_Surrogate_Dynamics/surrogate-dynamics/short_scripts/aligned.pdb',  # file to write the trajectory to
                in_memory=True
               ).run()
```
